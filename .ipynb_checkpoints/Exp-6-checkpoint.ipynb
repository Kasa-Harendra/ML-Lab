```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
x = df.iloc[:, 1:-1]
y = df['Species']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
rm = RandomForestClassifier(n_estimators=50, random_state=32)
```

```python
rm.fit(x_train, y_train)
```

```python
y_pred = rm.predict(x_test)
y_pred
```

```python
print('Accuracy: ', accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import *
cm = confusion_matrix(y_test, y_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=y_test.unique())
cm_display.plot()
print('Confusion matrix: \n', cm)
```

```python
print('Classification Report: ', classification_report(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 20))
plot_tree(rm.estimators_[0], feature_names = df.columns, class_names=y_test.unique(), filled=True)
plt.show()
```

```python
n_trees = len(rm.estimators_)
for i, tree in enumerate(rm.estimators_):
    plot_tree(tree, feature_names = df.columns, class_names=y_test.unique(), filled=True)
    plt.title(f'Tree {i}')
    plt.show()
```
