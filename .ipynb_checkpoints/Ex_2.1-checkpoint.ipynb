```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('New.csv')
df
```

```python
#1. feature Scaling 
from sklearn.preprocessing import StandardScaler, MinMaxScaler ,LabelEncoder, OneHotEncoder 
```

```python
# Initialize the StandardScaler 
scaler = StandardScaler() 
# Feature Scaling 
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']]) 
print(df)
```

```python
# Initialize the MinMaxScaler 
scaler = MinMaxScaler() 
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']]) 
print(df)
```

```python
#Categorical feature transformation 
# Encoding categorical data 
label_encoder = LabelEncoder() 
df['Purchased'] = label_encoder.fit_transform(df['Purchase']) 
df['Country'] = label_encoder.fit_transform(df['Country']) 
```

```python
#Unique values 
print("country unique values",df['Country'].unique()) 
print("purchased unique values",df['Purchased'].unique())
```

```python
#Remove Duplicate Records from Dataset 
df=df.drop_duplicates(keep='first') 
df
```

```python
#2.Data Visualization 
#Boxplot: 
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.boxplot(x='Age',y='Salary',data=df) 
plt.show()
```

```python
#Pait plot: 
sns.pairplot(df,hue='Purchased') 
plt.show()
```

```python
#line plot: 
sns.lineplot(x='Age',y='Salary',data=df,color='yellow') 
plt.show() 
```

```python
#Barplot: 
sns.barplot(x='Purchased',y='Country',data=df) 
plt.show()
```

```python
#Histogram: 
sns.histplot(x='Age',data=df,color='pink') 
plt.show() 
```

```python
#Scatterplot: 
sns.scatterplot(x='Age',y='Salary',data=df,color='green') 
plt.show()
```

```python
#regression plot: 
sns.regplot(x='Age',y='Salary',data=df,color='red')
plt.show()
```
