```python
import pandas as pd
import numpy as np
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
data = {
    "email":[
        "Win a free iphone now",
        "Meeting at 3 PM today",
        "Congratulations! You won a lottery",
        "Can you review my report?",
        "Claim your free vacation offer",
        "Let's schedule a meeting tomorrow",
        "Earn money fast and easy",
        "Are we still meeting for lunch?",
        "Get a free gift with your purchase",
        "This is your invoice for last month"
    ],
    "label":["spam","ham","spam","ham","spam","ham","spam","ham","spam","ham"]
}     
```

```python
df=pd.DataFrame(data)
df
```

```python
df["label"]=df["label"].map({"spam": 1,"ham": 0})
```

```python
vectorizer=TfidfVectorizer(stop_words="english")
x=vectorizer.fit_transform(df["email"])
y=df["label"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=MultinomialNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
new_email = ["You have won a free prize! Claim now"]
new_email_vectorized = vectorizer.transform(new_email)
```

```python
prediction = model.predict(new_email_vectorized)
print("New Email Prediction:", "Spam" if prediction[0] == 1 else "Ham")
```

```python
confusion_matrix=metrics.confusion_matrix(y_test,y_pred)
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=["spam","ham"])
cm_display.plot()
plt.show()
```

```python
print("Accuracy=",accuracy_score(y_test,y_pred))
print("Classification Report:",classification_report(y_test,y_pred))
```
