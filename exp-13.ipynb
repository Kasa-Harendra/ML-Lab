```python
from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Iris dataset
data = load_iris()
X = data.data
```

```python
# Fit GMM with EM (3 components for 3 classes)
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)
```

```python
# Predict cluster for each sample
labels = gmm.predict(X)
```

```python
# Plot with first two features
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='Set2')
plt.title("EM Algorithm via Gaussian Mixture Model on Iris")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
