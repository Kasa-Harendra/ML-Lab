```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("kmeans_blobs.csv")
data.head()
```

```python
x = data.iloc[:,:-1]
y = data['cluster']
```

```python
sns.scatterplot(x = "data1", y = "data2", data= data, hue="cluster")
plt.xlabel('data 1')
plt.ylabel('data 2')
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans
```

```python
y_pred=kmeans.fit_predict(x)
y_pred
data['cluster_pred']=y_pred
```

```python
kmeans.cluster_centers_
```

```python
X=kmeans.cluster_centers_[:,0]
Y=kmeans.cluster_centers_[:,1]
```

```python
df1=data[data.cluster_pred==0]
df2=data[data.cluster_pred==1]
df3=data[data.cluster_pred==2]
# Get the cluster labels (predicted clusters)
cluster_labels = kmeans.labels_
```

```python
new_point = np.array([[5.2, 10.3]])
predicted_cluster = kmeans.predict(new_point)[0]
print(f"The new point {new_point[0]} belongs to cluster: {predicted_cluster}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df1.data1, df1.data2, color='green',marker='o',label='cluster1')
plt.scatter(df2.data1, df2.data2, color='red', marker='d')
plt.scatter(df3.data1, df3.data2, color='blue' ,marker='s')
plt.scatter(X,Y,color='purple', marker='*',label='Centroid')
plt.scatter(new_point[0, 0], new_point[0, 1], c='black', s=150, marker='*', label='New Point')

plt.title('K-Means Clustering')
plt.xlabel('data1')
plt.ylabel('data2')
```
